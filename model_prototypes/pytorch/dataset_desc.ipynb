```python

test = df[df['fold'] == 'test']
test
```

```python
def get_heur_path(path):
    return path.replace('/more_cpp_2','/more_cpp_heuristic')

labels, heuristics = [],[]
act_i = 0
for i, row in tqdm(test.iterrows(),total=len(test)):
    # print(row['path'], get_heur_path(row['path']))
    if os.path.isfile(get_heur_path(row['path'])):
        # print('found match')
        base = pd.read_csv(row['path'],names=headers,delimiter=',')
        heur = pd.read_csv(get_heur_path(row['path']),names=headers,delimiter=',')
        labels.extend(list(base['branch_prob']))
        heuristics.extend(list(heur['branch_prob']))

    # if act_i > 100: break 
    # act_i +=1 
```

```python
print(metrics.mean_squared_error(labels,heuristics) )
print(metrics.mean_squared_error(labels,heuristics) **0.5)

print(metrics.mean_absolute_error(labels,heuristics) )
```
